```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = pd.period_range(pd.datetime.now(), periods=200, freq='d')
x = x.to_timestamp().to_pydatetime()
y = np.random.randn(200, 3).cumsum(0)
```

```python
plt.plot(x,y)
```

```python
plots = plt.plot(x,y)
plt.legend(plots, ('foo', 'bar', 'baz'), loc='best', framealpha=0.25,
           prop={'size': 'small', 'family': 'monospace'})
plt.gcf().set_size_inches(10,5)
plt.title('Random trends')
plt.xlabel('Date')
plt.ylabel('cum sum')
plt.grid('on')
plt.figtext(0.995, 0.01, '\u00a9 Acme Designs 2015', ha='right', va='bottom')
plt.tight_layout()
```

```python
fig = plt.figure(figsize=(10,5))
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.set_title('Main Axes with Insert Child Axes')
ax.plot(x, y[:,0])
ax.set_xlabel('Date')
ax.set_ylabel('cum sum')
ax = fig.add_axes([0.58, 0.55, 0.3, 0.3])
ax.plot(x, y[:,1], color='g')
ax.set_xticks([])
```

```python
fig, ax = plt.subplots()
plots = ax.plot(x, y, label='')
fig.set_size_inches(10,5)
ax.legend(plots, ('foo', 'bar', 'baz'), loc='best', framealpha=0.25,
           prop={'size': 'small', 'family': 'monospace'})
ax.set_title('Main Axes with Insert Child Axes')
ax.set_xlabel('Date')
ax.set_ylabel('cum sum')
ax.grid('on')
fig.text(0.995, 0.01, '\u00a9 Acme Designs 2015', ha='right', va='bottom')
fig.tight_layout()
```

```python
mpl.rcParams['lines.linewidth'] = 1
fig, axes = plt.subplots(
                nrows=3, ncols=1,
                sharex=True, sharey=True,
                figsize=(8,8))
label_data = zip(y.transpose(), ('foo', 'bar', 'baz'), ('b', 'g', 'r'))

for i, ld in enumerate(label_data):
    ax = axes[i]
    ax.plot(x, ld[0], label=ld[1], color=ld[2])
    ax.grid('on')
    ax.set_ylabel('cum sum')
    ax.legend(loc='upper left', framealpha=0.5, prop={'size': 'small'})
axes[-1].set_xlabel('Date')
fig.tight_layout()
```

```python
labels = ["Physics", "Chemistry", "Literature", "Peace"]
foo_data = [3, 6, 10, 4]
bar_width = 0.5
xlocations = np.array(range(len(foo_data))) + bar_width
plt.bar(xlocations, foo_data, width=bar_width)
plt.yticks(range(0,12))
plt.xticks(xlocations, labels)
# plt.xlim(0, xlocations[-1]+bar_width)
plt.title("Prizes won by Fooland")
gca = plt.gca()
gca.get_xaxis().tick_bottom()
gca.get_yaxis().tick_left()
plt.gcf().set_size_inches((10,5))
# gca.get_xlim()
```

```python
labels = ["Physics", "Chemistry", "Literature", "Peace"]
foo_data = [3, 6, 10, 4]
bar_data = [8, 3, 6, 1]

fig, ax = plt.subplots(figsize=(10,5))
bar_width = 0.4
xlocs = np.array(range(len(foo_data)))
plt.bar(xlocs-bar_width, foo_data, bar_width, color='#fde0bc', label='Fooland')
plt.bar(xlocs, bar_data, bar_width, color='peru', label='Barland')
ax.set_yticks(range(0,12))
ax.set_xticks(xlocs-bar_width/2)
ax.set_xticklabels(labels)
ax.yaxis.grid('on')
ax.legend(loc='best')
ax.set_ylabel('Number of prizes')
fig.suptitle('Prizes by country')
fig.tight_layout(pad=2)
fig.savefig('mpl_barchar_multi.png', dpi=200)
```

```python
labels = ["Physics", "Chemistry", "Literature", "Peace"]
foo_data = [3, 6, 10, 4]
bar_data = [8, 3, 6, 1]

fig, ax = plt.subplots(figsize=(10,5))
bar_width = 0.4
xlocs = np.array(range(len(foo_data)))
plt.barh(xlocs-bar_width, foo_data, bar_width, color='#fde0bc', label='Fooland')
plt.barh(xlocs, bar_data, bar_width, color='peru', label='Barland')
ax.set_xticks(range(0,12))
ax.set_yticks(xlocs-bar_width/2)
ax.set_yticklabels(labels)
ax.xaxis.grid('on')
ax.legend(loc='best')
ax.set_xlabel('Number of prizes')
fig.suptitle('Prizes by country')
fig.tight_layout(pad=2)
```

```python
labels = ["Physics", "Chemistry", "Literature", "Peace"]
foo_data = [3, 6, 10, 4]
bar_data = [8, 3, 6, 1]

fig, ax = plt.subplots(figsize=(10,5))
bar_width = 0.8
xlocs = np.array(range(len(foo_data))) - bar_width / 2
plt.bar(xlocs, foo_data, bar_width, color='#fde0bc', label='Fooland')
plt.bar(xlocs, bar_data, bar_width, color='peru', label='Barland', bottom=foo_data)
ax.set_yticks(range(0,18))
ax.set_xticks(xlocs)
ax.set_xticklabels(labels)
ax.yaxis.grid('on')
ax.legend(loc='best')
ax.set_ylabel('Number of prizes')
fig.suptitle('Prizes by country')
fig.tight_layout(pad=2)
```

```python
num_points = 100
gradient = 0.5
x = np.array(range(num_points))
y = np.random.randn(num_points) * 10 + x * gradient
fig, ax = plt.subplots(figsize=(10,5))
ax.scatter(x, y)
fig.suptitle('A Simple Scatterplot')
```

```python
num_points = 100
gradient = 0.5
x = np.array(range(num_points))
y = np.random.randn(num_points) * 10 + x * gradient
fig, ax = plt.subplots(figsize=(10,5))
colors = np.random.rand(num_points)
size = (2 + np.random.rand(num_points) * 18) ** 2
ax.scatter(x, y, s=size, c=colors, alpha=0.5)
fig.suptitle('Scatterplot with Colot and Size Specified')
```

```python
num_points = 100
gradient = 0.5
x = np.array(range(num_points))
y = np.random.randn(num_points) * 10 + x * gradient
fig, ax = plt.subplots(figsize=(10,5))
ax.scatter(x, y)
m, c = np.polyfit(x, y, 1)
ax.plot(x, m*x+c)
fig.suptitle('Scatterplot with Regression Line')
```

```python
import seaborn as sns
```

```python
data = pd.DataFrame({'dummy x': x, 'dummy y': y})
sns.lmplot('dummy x', 'dummy y', data, size=4, aspect=2)
plt.tight_layout()
plt.savefig('mpl_scatter_seaborn.png')
```

```python
sns.lmplot('dummy x', 'dummy y', data, size=6, aspect=2,
          scatter_kws={"color": "slategray"},
          line_kws={"linewidth": 2, "linestyle": "--", "color": "seagreen"},
          markers="D",
          ci=68)
plt.tight_layout()
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
g = sns.FacetGrid(tips, col="sex", size=6, aspect=1)
g.map(plt.scatter, "total_bill", "tip")
```

```python
g = sns.FacetGrid(tips, col="smoker", size=6, aspect=1)
g.map(plt.scatter, "total_bill", "tip")
```

```python
g = sns.FacetGrid(tips, col="day", size=5, aspect=1)
g.map(plt.scatter, "total_bill", "tip")
```

```python
pal = dict(Female='red', Male='blue')
g = sns.FacetGrid(tips, col="smoker", hue="sex", hue_kws={"marker": ["D", "s"]},
                  palette=pal, size=6, aspect=1)
g.map(plt.scatter, "total_bill", "tip", alpha=0.4)
g.add_legend()
```

```python
pal = dict(Female='red', Male='blue')
g = sns.FacetGrid(tips, col="smoker", row="time", 
                  hue="sex", hue_kws={"marker": ["D", "s"]},
                  palette=pal, size=6, aspect=1)
g.map(sns.regplot, "total_bill", "tip", scatter_kws={'alpha':0.6})
g.add_legend()
```

```python
pal = dict(Female='red', Male='blue')
sns.lmplot(x='total_bill', y='tip', hue="sex",
           markers=["D","s"], col="smoker", row="time", data=tips, palette=pal,
           size=6, aspect=1)
plt.tight_layout()
```
